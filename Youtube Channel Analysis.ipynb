```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/Global YouTube Statistics.csv' , encoding = 'latin1')
df
```

```python
df.info()
```

```python
for i in df:
    if df[i].dtypes == 'O':
        df[i] = df[i].str.lower()
df.head()
```

```python
df.drop_duplicates(inplace = True)
df.shape
```

```python
df.drop(['Abbreviation',
         'video_views_for_the_last_30_days',
         'subscribers_for_last_30_days',
         'created_month',
         'created_date'],
          inplace=True, axis=1)

df.isnull().sum() / len(df) * 100
```

```python
df.drop(df[df['Country'].isnull()].index , inplace = True)

df.isnull().sum() / len(df) * 100
```

```python
df.shape
```

```python
print(df['channel_type'].unique())
print('\n\n')
print(df['category'].unique())
```

```python
# Here we are have analyzed that category and channel_type column in df dataset are giving same values , that's why we are going to drop the category column
# and channel_type contains less null values than category
print("Category wise null values     :  ",df['category'].isnull().sum())
print("Channel type wise null values :  ",df['channel_type'].isnull().sum())
```

```python
df.drop('category' , axis = 1 , inplace = True)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df['yearly_earnings'] = (df['lowest_monthly_earnings'] + df['highest_monthly_earnings']) // 2
```

```python
plt.figure(figsize=[20, 10])
# Create the scatter plot
sns.scatterplot(x=df["subscribers"],  y=df["video views"], hue=df["channel_type"],  size=df["video views"])
plt.xlabel("Subscribers")
plt.ylabel("Views")
plt.title("Relationship between Views and Subscribers")
plt.show()
```

```python
plt.figure(figsize = [14 , 6])
sns.scatterplot(x = df['subscribers'] , y = df['yearly_earnings'] , hue = df['channel_type'] , size = df['yearly_earnings'])
plt.title('Yearly earnings as per subscribers')
plt.xlabel('Subscribers')
plt.ylabel('yearly earnings')
plt.show()
```

```python
plt.figure(figsize = [14 , 6])
sns.scatterplot(x = df['video views'] , y = df['yearly_earnings'] , hue = df['channel_type'] , size = df['yearly_earnings'])
plt.title('Yearly earnings as per video views')
plt.xlabel('video views')
plt.ylabel('yearly earnings')
plt.show()
```

```python
# Distribution of channels as according to it's counts
channel_type = df['channel_type'].value_counts().sort_values(ascending = False)
```

```python
plt.figure(figsize = [14 , 6])
plt.pie(channel_type.values , labels = channel_type.index , textprops = {'fontsize': 10})
plt.title('Channel type accroding to it\'s counts of channel ')
plt.show()
```

```python
top_20_channel_types = channel_type.head(20)
plt.figure(figsize = [14 , 7])
sns.countplot(data = df , x = 'channel_type' ,  order=top_20_channel_types.index)
plt.title('10 Most Popular Channel Categories')
plt.xlabel('channel_type')
plt.ylabel('Amount of Channels')
plt.xticks(rotation=45)
plt.show()
```

```python
df['channel_age'] = 2023 - df['created_year']
df['annual_subscriber_growth'] = df['subscribers']/df['channel_age']
```

```python
channel_type_growth = df.groupby('channel_type')['annual_subscriber_growth'].mean().reset_index()
channel_type_growth.sort_values(by='annual_subscriber_growth', ascending=False)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=channel_type_growth, x='channel_type', y='annual_subscriber_growth',)
plt.title('channel_type-Wise Average Subscribers Growth')
plt.xlabel('Channel Type')
plt.ylabel('Annual_Subscriber_Growth')
plt.xticks(rotation=45)
plt.show()
```

```python
category_earnings = df.groupby('channel_type')['yearly_earnings'].mean().reset_index()
sorted_category_earnings = category_earnings.sort_values(by = 'yearly_earnings' , ascending = False)
top_category_earnings = sorted_category_earnings.head(10)
```

```python
plt.figure(figsize = [14 , 7])
sns.barplot(data = top_category_earnings , x = 'channel_type' ,  y = 'yearly_earnings')
plt.xlabel('channel type')
plt.ylabel('yearly earnings')
plt.title('Yearly earnings Average of top channel types')
plt.show()
```

```python
df_geo_subscribers = df.groupby('Country')['subscribers'].mean().reset_index()
df_geo_subscribers.sort_values( 'subscribers', inplace=True, ascending=False )
df_geo_subscribers = df_geo_subscribers.head(20)
```

```python
plt.figure(figsize = [14 , 7])
sns.barplot(data  = df_geo_subscribers , x = 'Country' , y = 'subscribers')
plt.title('Average number of Subscribers by Country')
plt.xlabel('Country')
plt.ylabel('Average number of Subscribers')
plt.xticks(rotation = 90)
plt.show()
```

```python
import plotly.express as px

px.choropleth(df_geo_subscribers,
                    locations="Country",
                    locationmode='country names',
                    color='subscribers',
                    title="Average number of Subscribers by Country")
```

```python
df_geo_count = df.groupby('Country')['subscribers'].count().reset_index()
df_geo_count = df_geo_count.rename(columns = {'subscribers': "Count of channels"})

px.choropleth(df_geo_count,
                    locations="Country",
                    locationmode='country names',
                    color='Count of channels',
                    color_continuous_scale=px.colors.sequential.Emrld ,
                    title="Amount of channels per Country")
```
